```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.floridaresidentsdirectory.com/")
time.sleep(1)


btn = driver.find_element(By.XPATH, "/html/body/div[2]/div/div[1]/div/nav/button")
btn.click()
time.sleep(2)


number_btn = driver.find_element(By.XPATH, "//*[@id='navbarSupportedContent']/ul/li[4]/a")
number_btn.click()
time.sleep(2)


phone_btn = driver.find_element(By.XPATH, "/html/body/div[2]/div/div[3]/div/div[2]/div/div/div[1]/a[1]")
phone_btn.click()
time.sleep(2)


elements = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.XPATH, "//a[@class='btn btn-block btn-white col-green']"))
)


product_urls = []


for element in elements:
    url = element.get_attribute("href")
    if url:
        product_urls.append(url)


for product_url in product_urls:
    driver.get(product_url)
    time.sleep(2)  
    
    
    product_elements = driver.find_elements(By.XPATH, "//div[@class='similar']/a")
    
    
    product_link_list = []
    
    
    for product_element in product_elements:
        product_link = product_element.get_attribute("href")
        if product_link:
            product_link_list.append(product_link)
    
   
    for product_link in product_link_list:
        driver.get(product_link)
        time.sleep(2)  
```
