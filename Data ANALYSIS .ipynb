```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("P1-SuperStoreUS-2015 - Orders.csv")
```

```python
data.head(2)
```

# missing value find code 

```python
# missing value find code 
data.isnull().sum()
```

# missing value fill code 

```python
# missing value fill code 
data["Product Base Margin"].fillna(data["Product Base Margin"].mean(),inplace=True)
```

```python
# data check code
data.shape
```

# Order Priority

```python
data["Order Priority"].value_counts()
```

```python
# colum check code 
data["Order Priority"].unique()
```

```python
# replace colum code 
data ["Order Priority"] = data ['Order Priority'].replace("Critical ","Critical")
```

```python
# garaf size low code 
plt.figure(figsize=(5,4))
# garaf tittle code 
plt.title("count of Order Priority")
# garaf banaye k code 
sns.countplot(x="Order Priority", data=data)
# garaf save code
plt.savefig("count of Order Priority.jpg")
plt.show()
```
