```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.floridaresidentsdirectory.com/search/phone/239/212/")
time.sleep(1)


product_elements = driver.find_elements(By.XPATH, "//div[@class='similar']/a")
product_link_list = []

for product_element in product_elements:
    product_link = product_element.get_attribute("href")
    if product_link:
        product_link_list.append(product_link)

# Open a CSV file for writing
with open('data_3_apr_24.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Name', 'Location', 'Age', 'Gender', 'Phone', 'Email', 'URL']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for product_link in product_link_list:
        driver.get(product_link)
        time.sleep(2)

        data = {}
        try:
            name_element = driver.find_element(By.XPATH, "//*[@id='voter-record-header']/h1/span")
            data['Name'] = name_element.text
        except Exception as e:
            data['Name'] = "Not Found"

        try:
            location_element = driver.find_element(By.XPATH, "//*[@id='person-data-wrapper']/a[1]")
            data['Location'] = location_element.text
        except Exception as e:
            data['Location'] = "Not Found"

        try:
            age_element = driver.find_element(By.XPATH, "//*[@id='person-data-wrapper']/div[2]/div/div[1]/p[2]")
            data['Age'] = age_element.text
        except Exception as e:
            data['Age'] = "Not Found"

        try:
            gender = driver.find_element(By.XPATH, "//span[@itemprop='gender']")
            data['Gender'] = gender.text
        except Exception as e:
            data['Gender'] = "Not Found"

        try:
            phone_click = driver.find_element(By.XPATH, "//span[@itemprop='telephone']")
            phone_click.click()
            phone = driver.find_element(By.XPATH, "//*[@id='person-data-wrapper']/div[2]/div/div[1]/p[3]/span/span/span[1]/a")
            data['Phone'] = phone.text
        except Exception as e:
            data['Phone'] = "Not Found"

        try:
            email_click = driver.find_element(By.XPATH, "//span[@class='blts']")
            email_click.click()
            email = driver.find_element(By.XPATH, "//*[@id='person-data-wrapper']/div[2]/div/div[2]/p[3]/span/span[1]/a")
            data['Email'] = email.text
        except Exception as e:
            data['Email'] = "Not Found"

        data['URL'] = product_link

        writer.writerow(data)
        print("Data written to CSV:", data)
```
